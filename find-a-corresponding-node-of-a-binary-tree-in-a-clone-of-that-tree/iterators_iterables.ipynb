```python
# list is iterable because we can loop over it
nums = [1,2,3]
for num in nums:
    print(num)
```

```python
# many things can be looped over (iterable)
# iterables have special method __iter__()
# dunder special or magic method
print(dir(nums))
```

```python
# they know where they are and how to get the next value
# dunder next but list doesn't have next so it's not an iterator
print(next(nums)) #iterable but not iterator (no __next__())
```

```python
# i_nums = nums.__iter__() # makes an iterator for our list
i_nums = iter(nums) # same as above
print(i_nums)
print(dir(i_nums)) # this has next and iter
```

```python
# print(next(i_nums)) # it remembers where it is so next call should be 2
# print(next(i_nums)) # it remembers where it is so next call should be 3
# print(next(i_nums)) # it remembers where it is so next call should be None
# print(next(i_nums)) # it remembers where it is so this is exception
# for loop will handle this exception and ignore it
# same as above
while True:
    try:
        item = next(i_nums)
        print(item)
    except StopIteration:
        break
# iterators only go forward you can start over with a new iterator
```

```python
import random as r
# recreating range()
class MyRange:
    def __init__(self, start: int, end: int = None, increment: int = 1):
        self.start = start
        self.end = end
        if end is None:
            self.start = 0
            self.end = start
        self.current = start
        self.increment = increment

    def __iter__(self):
        return self # since we are defining next

    def __next__(self):
        curr = self.current
        if curr >= self.end:
            raise StopIteration
        self.current = self.current + self.increment
        return curr

def get_random_arr(start: int = 0, end: int = 100, count: int = 100):
    return [r.randint(start, end) for i in range(count)]
for i,j,k in zip(get_random_arr(),get_random_arr(),get_random_arr(1,100,100)): # randint inclusive both range
    assert list(range(i,j,k)) == list(MyRange(i,j,k))
# for i in MyRange(3,30,10):
#     print(i)
```

```python
# now for generators
# they yeild value until next is call
# next and iter are created automatically
# not class because we don't need instances
def my_range(start: int, end: int = None, increment: int = 1):
    _start = start
    _end = end
    if not end:
        _start = 0
        _end = start
    current = _start
    while current < end:
        yield current # yield is not return
        current += increment

def get_random_arr(start: int = 0, end: int = 100, count: int = 100):
    return [r.randint(start, end) for i in range(count)]
for i,j,k in zip(get_random_arr(),get_random_arr(),get_random_arr(1,100,100)): # randint inclusive both range
    assert list(range(i,j,k)) == list(my_range(i,j,k))

# for i in my_range(3,30, 10):
#     print(i)
```
